```python
import nidaqmx
import numpy as np
import matplotlib.pyplot as plt
from collections import deque
from IPython.display import display, clear_output
import csv

# Parmetros de adquisicin
puerto= "Dev3/ai0"  # Cambia esto segn el canal de tu DAQ
fmuestreo = 2000  # Frecuencia de muestreo en Hz
t_ventana = 2000  # Cantidad de muestras a mostrar en la grfica
duracion = 60  # Duracin de la adquisicin en segundos

# Crear estructura de datos para almacenar la seal en tiempo real
emg_buffer = deque([0] * t_ventana, maxlen=t_ventana)  # Guarda los ltimos datos capturados
tiempo_buffer = deque(np.linspace(0, (t_ventana - 1) / fmuestreo,t_ventana), maxlen=t_ventana)  # Eje X desde 0

# Archivo de salida para guardar los datos
filename = "emg_data.csv"
with open(filename, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Tiempo (s)", "Voltaje (V)"])  # Escribir encabezado

    # Configurar la tarea de adquisicin
    with nidaqmx.Task() as task:
        task.ai_channels.add_ai_voltage_chan(puerto)
        task.timing.cfg_samp_clk_timing(fmuestreo)

        print("Adquiriendo datos en tiempo real... Presiona 'Ctrl + C' para detener.")

        # Inicializar la figura
        plt.ion()  # Modo interactivo
        fig, ax = plt.subplots(figsize=(10, 4))
        line, = ax.plot(tiempo_buffer, emg_buffer, label="Seal EMG")
        ax.set_xlabel("Tiempo (s)")
        ax.set_ylabel("Voltaje (V)")
        ax.set_title("Seal EMG en Tiempo Real")
        ax.legend()
        ax.grid(True)
        ax.set_ylim([-1, 5])  # Ajuste de voltaje de -1V a 1V

        # Bucle de adquisicin en tiempo real
        try:
            total_samples = int(duracion * fmuestreo)
            for i in range(0, total_samples, t_ventana):
                # Leer mltiples muestras a la vez
                emg_data = task.read(number_of_samples_per_channel=t_ventana)

                if isinstance(emg_data, list):  # Confirmar que es una lista
                    emg_buffer.extend(emg_data)

                    # Guardar los datos en el archivo
                    for j in range(len(emg_data)):
                        tiempo = (i + j) / fmuestreo
                        writer.writerow([tiempo, emg_data[j]])

                    # Actualizar grfico en tiempo real
                    line.set_ydata(emg_buffer)
                    clear_output(wait=True)
                    display(fig)
                    plt.pause(0.01)

        except KeyboardInterrupt:
            print("Adquisicin detenida por el usuario.")

        plt.ioff()
        plt.show()

print(f"Los datos se han guardado exitosamente en '{filename}'.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import butter, filtfilt, freqz

#  Cargar los datos desde el archivo CSV
filename = "emg_data_modified.csv"
data = pd.read_csv(filename)
tiempos = data['Tiempo (s)'].values
voltajes = data['Voltaje (V)'].values

# Parmetros del filtro
fs = 2000  # Frecuencia de muestreo en Hz (asegrate de que sea la misma que usaste para adquirir la seal)
bajaf = 20  # Frecuencia de corte para el filtro pasa altas (en Hz)
altaf = 450  # Frecuencia de corte para el filtro pasa bajas (en Hz)

# Funcin para disear un filtro Butterworth
def butter_filter(fcorte, fs, order=4, filter_type='low'):
    nyquist = 0.5 * fs
    normal_fcorte = fcorte / nyquist
    b, a = butter(order, normal_fcorte, btype=filter_type, analog=False)
    return b, a

# Filtro Pasa Altas
b_alto, a_alto = butter_filter(bajaf, fs, order=4, filter_type='high')
pasa_altas = filtfilt(b_alto, a_alto, voltajes)

# Filtro Pasa Bajas
b_baja, a_baja = butter_filter(altaf, fs, order=4, filter_type='low')
filtrada = filtfilt(b_baja, a_baja, pasa_altas)

# Graficar seal original y filtrada
plt.figure(figsize=(17, 10))

plt.subplot(2, 1, 1)
plt.plot(tiempos, voltajes, label="Seal EMG Original", color='gray')
plt.title("Seal EMG Original")
plt.xlabel("Tiempo (s)")
plt.ylabel("Voltaje (V)")
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(tiempos, filtrada, label="Seal EMG Filtrada", color='blue')
plt.title("Seal EMG Filtrada (Pasa Altas y Pasa Bajas)")
plt.xlabel("Tiempo (s)")
plt.ylabel("Voltaje (V)")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()


# Diagrama de Bode para el filtro pasa altas (20 Hz)
w_high, h_high = freqz(b_alto, a_alto, worN=8000, fs=fs)  # Reducimos worN para menos detalle
plt.figure(figsize=(15, 6))

plt.subplot(2, 1, 1)
plt.plot(w_high, 20 * np.log10(abs(h_high)))
plt.axvline(bajaf, color='red', linestyle='--', label=f'Frecuencia de corte = {bajaf} Hz')
plt.title('Diagrama de Bode - Filtro Pasa Altas (20 Hz)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud (dB)')
plt.xlim(0, 100)  # Limitamos el rango de frecuencias para mejor visualizacin
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
angles_high = np.unwrap(np.angle(h_high))
plt.plot(w_high, angles_high)
plt.axvline(bajaf, color='red', linestyle='--', label=f'Frecuencia de corte = {bajaf} Hz')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase (radianes)')
plt.xlim(0, 100)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

# Diagrama de Bode para el filtro pasa bajas (450 Hz)
w_low, h_low = freqz(b_baja, a_baja, worN=8000, fs=fs)  # Reducimos worN para menos detalle
plt.figure(figsize=(15, 6))

plt.subplot(2, 1, 1)
plt.plot(w_low, 20 * np.log10(abs(h_low)))
plt.axvline(altaf, color='red', linestyle='--', label=f'Frecuencia de corte = {altaf} Hz')
plt.title('Diagrama de Bode - Filtro Pasa Bajas (450 Hz)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud (dB)')
plt.xlim(0, 1000)  # Limitamos el rango de frecuencias para mejor visualizacin
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
angles_low = np.unwrap(np.angle(h_low))
plt.plot(w_low, angles_low)
plt.axvline(altaf, color='red', linestyle='--', label=f'Frecuencia de corte = {altaf} Hz')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Fase (radianes)')
plt.xlim(0, 1000)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
from numpy import hamming

# Cargar la seal desde el CSV

df = pd.read_csv("emg_filtrada.csv")

tiempos = df["Tiempo (s)"].values
voltajes = df["Voltaje (V)"].values

# Detectar picos (contracciones musculares)
picos, _ = find_peaks(voltajes, height=np.mean(voltajes) + 3 * np.std(voltajes), distance=300)

# Configurar ventana Hamming (200 ms)
dt = tiempos[1] - tiempos[0]
n = int(0.2 / dt)
if n % 2 == 0:
    n += 1
ventana = hamming(n)

# Aplicar ventana a los segmentos alrededor de cada pico
ventanas_aplicadas = []
tiempos_ventanas = []
for pico in picos:
    ini = pico - n // 2
    fin = pico + n // 2 + 1
    if ini >= 0 and fin <= len(voltajes):
        segmento = voltajes[ini:fin] * ventana
        ventanas_aplicadas.append(segmento)
        tiempos_ventanas.append(tiempos[ini:fin])

# Graficar la seal completa y superponer las ventanas aplicadas


plt.figure(figsize=(12, 4))
plt.plot(tiempos, filtrada, label='Seal EMG Filtrada', color='blue')


for t_vent, v_vent in zip(tiempos_ventanas, ventanas_aplicadas):
    plt.plot(t_vent, v_vent, color='orange', linewidth=2, alpha=0.7)

plt.title("Seal EMG con Ventanas de Hamming Aplicadas")
plt.xlabel("Tiempo (s)")
plt.ylabel("Voltaje (V)")
plt.grid(True)
plt.legend(["Seal EMG", "Ventanas Hamming"])
plt.tight_layout()
plt.show()

print("Cantidad de ventanas aplicadas:", len(ventanas_aplicadas))

```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import find_peaks, welch
from numpy import hamming

# Cargar la seal
df = pd.read_csv("emg_filtrada.csv")
tiempos = df["Tiempo (s)"].values
voltajes = df["Voltaje (V)"].values

# Detectar picos (contracciones)
picos, _ = find_peaks(voltajes, height=np.mean(voltajes) + 3 * np.std(voltajes), distance=300)

# Configurar ventana Hamming (200 ms)
dt = tiempos[1] - tiempos[0]
fs = 1 / dt
n = int(0.2 / dt)
if n % 2 == 0:
    n += 1
ventana = hamming(n)

# Crear seal con ventanas (todo lo dems en cero)
ventana_emg = np.zeros_like(voltajes)
for pico in picos:
    ini = pico - n // 2
    fin = pico + n // 2 + 1
    if ini >= 0 and fin <= len(voltajes):
        ventana_emg[ini:fin] = voltajes[ini:fin] * ventana

# Calcular la Densidad Espectral de Potencia (PSD) con Welch
frecuencias, psd = welch(ventana_emg, fs=fs, window='hamming', nperseg=n)

# Graficar la PSD
plt.figure(figsize=(10, 4))
plt.semilogy(frecuencias, psd, color='green')
plt.title("Densidad espectral de potencia (PSD) de la seal EMG aventanada")
plt.xlabel("Frecuencia (Hz)")
plt.ylabel("Densidad de Potencia (V/Hz)")
plt.grid(True, which='both', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from scipy.signal import welch
import matplotlib.pyplot as plt

def calcular_frecuencia_mediana(signal, fs):
    f, Pxx = welch(signal, fs=fs)
    energia_acumulada = np.cumsum(Pxx)
    energia_total = energia_acumulada[-1]
    f_mediana = f[np.where(energia_acumulada >= energia_total / 2)[0][0]]
    return f_mediana

def calcular_frecuencia_media(signal, fs):
    f, Pxx = welch(signal, fs=fs)
    f_media = np.sum(f * Pxx) / np.sum(Pxx)
    return f_media

def calcular_estadisticos(signal):
    return {
        "Media": np.mean(signal),
        "Mediana": np.median(signal),
        "Desviacin estndar": np.std(signal),
        "Mximo": np.max(signal),
        "Mnimo": np.min(signal)
    }

# -------------------------------
# Cargar archivo CSV
# -------------------------------
df = pd.read_csv("senal_emg_.csv")  # Asegrate que el nombre del archivo sea correcto
senal = df["Voltaje (V)"].values
fs = 2000  # Frecuencia de muestreo en Hz

# -------------------------------
# Seleccionar ventanas
# -------------------------------
duracion_ventana = 2  # segundos
muestras_ventana = int(fs * duracion_ventana)

ventanas = [senal[i:i + muestras_ventana] for i in range(0, len(senal), muestras_ventana)
            if len(senal[i:i + muestras_ventana]) == muestras_ventana]

# -------------------------------
# Calcular estadsticas y frecuencias
# -------------------------------
resultados = []
for ventana in ventanas:
    stats = calcular_estadisticos(ventana)
    stats["Frecuencia mediana"] = calcular_frecuencia_mediana(ventana, fs)
    stats["Frecuencia media"] = calcular_frecuencia_media(ventana, fs)
    resultados.append(stats)

# -------------------------------
# Resultados en DataFrame
# -------------------------------
df_resultados = pd.DataFrame(resultados)

# Mostrar primeras y ltimas filas
print(df_resultados.head())
print(df_resultados.tail())

# -------------------------------
# Calcular y mostrar media y desviacin estndar de las frecuencias medias
# -------------------------------
media_frec_media = df_resultados["Frecuencia media"].mean()
std_frec_media = df_resultados["Frecuencia media"].std()


```

```python
# -------------------------------
# Tomar la primera y ltima ventana
# -------------------------------
ventana_inicial = ventanas[0]
ventana_final = ventanas[-1]

# Calcular estadsticas
media_inicial = np.mean(ventana_inicial)
media_final = np.mean(ventana_final)

std_inicial = np.std(ventana_inicial, ddof=1)  # ddof=1 para desviacin estndar muestral
std_final = np.std(ventana_final, ddof=1)

n = len(ventana_inicial)  # Nmero de muestras por ventana

# -------------------------------
# Calcular t con frmula manual
# -------------------------------
numerador = media_inicial - media_final
denominador = np.sqrt((std_inicial**2 / n) + (std_final**2 / n))
t_manual = numerador / denominador

print("\n Prueba de hiptesis entre primera y ltima ventana (t calculado manualmente):")
print(f"Media inicial: {media_inicial:.4f}")
print(f"Media final:   {media_final:.4f}")
print(f"Desv. inicial: {std_inicial:.4f}")
print(f"Desv. final:   {std_final:.4f}")
print(f"t = {t_manual:.4f}")
```

```python
from scipy.stats import ttest_ind
# -------------------------------
# Prueba de hiptesis con t-test
# -------------------------------

# Selecciona la primera y ltima ventana como arreglos de datos
datos_inicial = ventanas[0]
datos_final = ventanas[-1]

# Realiza el test t de muestras independientes
t_stat, p_value = ttest_ind(datos_inicial, datos_final, equal_var=False)  # Welch's t-test

# Muestra los resultados
print(f"\nEstadstico t calculado: {t_stat:.4f}")
print(f"Valor-p: {p_value:.4f}")

# Interpretacin simple
alpha = 0.05
if p_value < alpha:
    print(" Hay evidencia significativa de una diferencia entre las ventanas (rechazamos H0).")
else:
    print(" No hay evidencia significativa de diferencia por lo que no se rechaza la hipotesis nula (H0).")
```
